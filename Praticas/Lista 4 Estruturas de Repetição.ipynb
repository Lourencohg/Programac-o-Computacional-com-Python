1. Faça um programa em Python que receba um número inteiro positivo e mostre o valor do seu fatorial.

```python
Num = int(input("Entre com o valor positivo a ser fatorado: "))
Fatorial = Num
Multi = Num - 1

if(Num < 0):
  print("Valores negativos não atendem as diretrizes do sistema")

while Multi > 0 :
  Fatorial = Fatorial * Multi
  Multi = Multi - 1
  if Multi == 0:
    print(Fatorial)
```

2. Faça um programa em Python que receba um número inteiro positivo e determine se esse número é primo.

```python
Num = int(input("Entre com o valor do número a ser testado: "))
Divis = 0

for i in range(1 , Num + 1 ):
  if Num % i == 0:
    Divis += 1

if Divis == 2:
  print(f"{Num} é um némero primo")
else:
  print(f"{Num} não é um número primo")
```

3. Faça um programa em Python que mostre todos os números primos no intervalo (fechado) de 50 a 150.

```python
Inicio = 50
Fim = 150

print(f"Números primos no intervalo de {Inicio} a {Fim}: ")

for num in range(Inicio, Fim + 1): # loop para cada número do intervalo
    e_primo = True # assumir primo atpe ser provado contrario

    for i in range(2, num):
        if num % i == 0:
            e_primo = False
            break

    #Se continuou True o número é primo
    if e_primo:
        print(num, end=" ")
```

4. Faça um programa em Python que receba um número inteiro positivo e mostre todos os números no intervalo (fechado) de 1 a 100 que são divisíveis por esse número entrado.

5. Faça um programa em Python que receba um valor inteiro positivo $N$ e mostre os $N$ primeiros números primos (começando por 2).

```python
Num = int(input("Entre com o valor de N: "))
n_primos = 0
numero_atual = 2 # o primeiro número primo é 2

while n_primos < Num:

    e_primo = True

    for i in range(2, numero_atual):
        if numero_atual % i == 0:
            e_primo = False  #não é primo
            break

    if e_primo:
        print(numero_atual, end=" ")
        n_primos += 1  #

    numero_atual += 1
```
