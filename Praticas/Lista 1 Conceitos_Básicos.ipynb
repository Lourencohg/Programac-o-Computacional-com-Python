1. Faça um programa que receba duas notas e os pesos delas como entrada e apresenta a média ponderada.

```python
Nota1 = int(input("Escreva a primeira nota:"))
Nota2 = int(input("Escreva a segunda nota:"))

Peso1 = int(input("Escreva o peso da primeira nota:"))
Peso2 = int(input("Escreva o peso da segunda nota:"))

Medp = (Nota1*Peso1 + Nota2*Peso2)/(Peso1 + Peso2)
print(Medp)
```

2.   Faça um programa que calcula o Índice de Massa Corporal (IMC) de uma pessoa a partir de sua altura em metros e o seu peso em Kg, como dado na fórmula: $IMC = Peso / (Altura *Altura)$

```python
print("Para calcular o IMC por favor insira os dados abaixo:")
Peso = int(input("Peso da pessoa em Kg:"))
Alt = float(input("Altura da pessoa em metros:"))

IMC = Peso/(Alt**2)
print(f"O Índice de Massa é de: {IMC}")
```

3.   Faça um algoritmo para calcular a média aritmética M entre duas notas de um aluno e mostrar sua situação, que pode ser aprovado (M ≥ 7), reprovado (M < 4) e AF (4 ≤ M < 7). Se o aluno ficar de AF, entre com a nota da AF e mostre a média e o resultado final.

```python
Nota1 = int(input("Escreva a primeira nota:"))
Nota2 = int(input("Escreva a segunda nota:"))

Medp = (Nota1 + Nota2)/(2)
print(f"A nota final do aluno é {Medp}")

P_AF = False
if Medp >= 7:
  print("Aluno foi aprovado")
elif Medp >= 4:
  print("Aluno está de AF")
  P_AF = True
else:
  print("Aluno foi reprovado")

if P_AF == True:
  Nota_AF = int(input("Escreva a nota da AF:"))
  Medp = (Medp + Nota_AF)/2
  print(f"A nota final do aluno é {Medp}")

if Medp >= 5:
  print("Aluno foi aprovado por AF")
else:
  print("Aluno foi reprovado")
```

4.   Joãozinho quer calcular e mostrar a quantidade de litros de combustível gastos em uma viagem, ao utilizar um automóvel que faz 12 KM/L. Para isso, ele gostaria que você o auxiliasse através de um simples algoritmo. Para efetuar o cálculo, deve-se fornecer como entrada o tempo gasto na viagem (em horas) e a velocidade média durante a mesma (em KM/H). Assim, pode-se obter distância percorrida e, em seguida, calcular quantos litros seriam necessários.

```python
Eficiência = 12

T = float(input("Escreva o tempo gasto na viagem em horas:"))
V = int(input("Escreva a velocidade média durante a viagem em KM/H:"))

Distancia = V * T

Gasto = Distancia / Eficiência
print(f"Foi gasto {Gasto:.3f} litros de combustível")
```
