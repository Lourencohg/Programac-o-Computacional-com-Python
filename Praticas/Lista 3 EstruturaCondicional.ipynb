**3**. Construa um algoritmo que dado quatro valores (A, B, C e D), o algoritmo imprima o maior e o menor valor.

```python
A = int(input("Entre com o valor de A: "))
B = int(input("Entre com o valor de B: "))
C = int(input("Entre com o valor de C: "))
D = int(input("Entre com o valor de D: "))

if(A > B):
  Maior1 = A
  Menor1 = B
else:
  Maior1 = B
  Menor1 = A
#Comparação entre A e B para descobrir qual o menor e maior entre os dois

if(C > D):
  Maior2 = C
  Menor2 = D
else:
  Maior2 = D
  Menor2 = C
#Comparação entre C e D para descobrir qual o menor e maior entre os dois

if(Maior1 > Maior2):
  print(f"{Maior1} é o maior valor")
else:
  print(f"{Maior2} é o maior valor")
#Comparação feita entre os maiores valores obtidos anteriosmente

if(Menor1 < Menor2):
  print(f"{Menor1} é o menor valor")
else:
  print(f"{Menor2} é o menor valor")
#comparação feita entre os menores valores obtidos anteriormente
```

**4.** Escreva um algoritmo que leia um número e informe se ele é divisível por 10, por 5 ou por 2 ou se não é divisível por nenhum deles.
